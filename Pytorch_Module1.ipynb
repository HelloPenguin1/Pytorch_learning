Pytorch Simple Linear NN

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype= torch.float32) #outer bracket represents that this data is a subset of the entire dataset. It is batch, and the inner data represents samples
times = torch.tensor([[6.90], [12.11], [16.42], [26.12]], dtype= torch.float32)
```

```python
model = nn.Sequential(nn.Linear(1,1))  #linear layer that takes in 1 input and outputs 1
```

```python
loss_func = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), #weights and biases
                      lr=0.01) #learning rate
```

Training in pytorch

How does training loop work in PyTorch?

For each epoch:
- clear out the gradients (necessary in pytorch else all previosuly computed gradients will accumulate)
- make predictions
- calc the loss
- calculate adjustments
- update the model

```python
inputs = torch.tensor([[5.0], [7.0], [8.0], [9.0]], dtype= torch.float32)
```

```python
for epoch in range(100):
  #Reset optimizer
  optimizer.zero_grad()
  #Make predictions
  outputs = model(distances)
  #calculate the loss
  loss = loss_func(outputs, times)
  #Calc adjustments
  loss.backward()
  #update model
  optimizer.step()
```

inference

```python
with torch.no_grad():
  test_dist = torch.tensor([25.0], dtype = torch.float32)
  prediction = model(test_dist)
  print(f"Predicted Time: {prediction.item():.2f}")
```
