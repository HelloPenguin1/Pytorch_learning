```python
import torch
```

```python
a = [[1,2,3],[4,5,6]]
a_t = torch.tensor(a, dtype=torch.float32)
```

```python
a_t
```

Mean of each column, set dim=0 to collapse rows, one value per column

```python
col_mean = torch.mean(a_t, dim=0)
print(col_mean)
```

Mean of each row, set dim=1 to collapse columns, one value per row

```python
row_mean = torch.mean(a_t, dim=1)
print(row_mean)
```

Mean of entire matrix

```python
entire_mean = torch.mean(a_t)
print(entire_mean)
```
